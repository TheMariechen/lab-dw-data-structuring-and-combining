**Summary of different merging fucntions 
=> concat used primarily to merge vertically (same column names no matter the order in df), while others primarily used to merge horizontally <=**

```python
import pandas as pd

url1 = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv'
df1 = pd.read_csv(url1)

url2 = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv'
df2 = pd.read_csv(url2)

url3 = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv'
df3 = pd.read_csv(url3)
```

```python
df1
```

```python
df2
```

```python
df3
```

**Concat works based on same column NAME not position --> do it does not matter where the columns are in the df, just that the name is the exact same!**

```python
wrong = pd.concat([df1, df2, df3]) # concat it as it is to see straight away which columns get duplicated  
```

```python
wrong # now we see that the columns State and Gender from df3 have been added. They need to be renamed to ST and GENDER as well 
```

```python
# Renaming State and Gender in df3 

df3.rename(columns={'State': 'ST', 'Gender': 'GENDER'}, inplace=True)
df3
```

```python
# Now they can be correctly concatinated 
df = pd.concat([df1, df2, df3]) 
df # looks correct 
```

```python
# Now data cleaning
# 1. homogenous column names 
df.columns = [column.lower() for column in df.columns] 
df.columns = [column.replace(' ', '_') for column in df.columns] 
df.rename(columns= {'st': 'state'}, inplace=True)
df.columns
```

```python
# 2. Cleaning invalid values 
df['gender'].unique()
df["gender"] = df["gender"].apply(lambda x: "F" if x in ("Femal", "female") else "M" if x == "Male" else x)
df['gender'].unique()
```

```python
df['state'].unique() # array(['Washington', 'Arizona', 'Nevada', 'California', 'Oregon', 'Cali', 'AZ', 'WA', nan], dtype=object)
df['state'] = df['state'].map({'WA': 'Washington', 'AZ': 'Arizona', 'Nevada': 'Nevada', 'Cali': 'California', 'Oregon':'Oregon'})
df['state'].unique() 
```

```python
df['education'].unique()
df['education'] = df['education'].apply(lambda name: "Bachelor" if name == "Bachelors" else name)
df['education'].unique()
```

```python
df['customer_lifetime_value'] = df['customer_lifetime_value'].str.replace('%', '')
df['customer_lifetime_value']
```

```python
df['vehicle_class'].unique() 
df['vehicle_class'] = df['vehicle_class'].map({'Four-Door Car': 'Four-Door Car', 'Two-Door Car': 'Two-Door Car', 'SUV': 'SUV', 'Luxury SUV': 'Luxury', 'Sports Car':'Luxury', 'Luxury Car': 'Luxury'})
df['vehicle_class'].unique()
```

```python
# Formatting data types 
df.dtypes
```

```python
df['customer_lifetime_value'] = df[['customer_lifetime_value']].applymap(float)
df.dtypes
```

```python
df['number_of_open_complaints'].unique()
df['number_of_open_complaints'] = df['number_of_open_complaints'].apply(lambda x: x.split('/')[1] if type(x) == str else x)
df['number_of_open_complaints'].unique()
# this bad looking array with multiple times the same number is resolved by transforming the column to float below! 
```

```python
df['number_of_open_complaints'].value_counts() 
```

```python
df['number_of_open_complaints'] = df[['number_of_open_complaints']].applymap(float)
```

```python
df['number_of_open_complaints'].value_counts()
```

```python
df.dtypes
```

```python
# 4. Dropping duplicates 
df.duplicated().sum() 
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.duplicated().sum() 
```

```python
# 5. Missings 
df.isna().any()
```

```python
df.isna().sum()
```

```python
# After dropping a lot of rows, not that many columns with a problematic number of missings remain: state, gender and customer_lifetime_value.
# We cant really fill any values for the first two string value columns, but we can fill the missings of customer_lifetime_value with the mean 
# (and we could drop state altogether because 5424/9131 rows with missings is more than half of all rows!
```

```python
df
```

```python
# filling the missings of customer_lifetime_value with the column mean
df["customer_lifetime_value"].fillna(df["customer_lifetime_value"].mean(), inplace = True)
```

```python
df.isna().sum() # boom, zero missings on customer_lifetime_value
```
